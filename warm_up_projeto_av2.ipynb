## 10. Crie um gráfico que mostre a distribuição de arrecadação dos 10 filmes de maior bilheteria mundial.

```python
import matplotlib.pyplot as plt

class GraficoBarras:
    def __init__(self):
        self.x = None
        self.y = None

    def criar_grafico(self, x, y):
        self.x = x
        self.y = y

    def configura_titulo_e_rotulos(self, titulo=None, rotulo_x=None, rotulo_y=None):
        if titulo is not None:
            plt.title(titulo)
        if rotulo_x is not None:
            plt.xlabel(rotulo_x)
        if rotulo_y is not None:
            plt.ylabel(rotulo_y)

    def exibir_grafico(self):
        plt.bar(self.x, self.y)
        # rotate the x-axis labels by 90 degrees
        plt.xticks(rotation=90)
        plt.show()
```

```python
lista_nomes = df_top_10_bilheteria['title'].to_list()
print(lista_nomes)

lista_bilheteria = df_top_10_bilheteria['worldwide_gross'].to_list()
print(lista_bilheteria)
```

```python
# Exemplo de uso da classe GraficoBarras
grafico_barras_top_10 = GraficoBarras()
categorias = lista_nomes
valores = lista_bilheteria
grafico_barras_top_10.criar_grafico(categorias, valores)
grafico_barras_top_10.exibir_grafico()
```

## 11. Crie um gráfico que mostre ao percentual de partição dos 10 filmes de maior bilheteria mundial.


```python
import matplotlib.pyplot as plt

class GraficoPizza:
    def __init__(self):
        self.labels = None
        self.valores = None

    def criar_grafico(self, labels, valores):
        self.labels = labels
        self.valores = valores

    def configura_titulo(self, titulo=None):
      if titulo is not None:
        plt.title(titulo)

    def exibir_grafico(self):
        plt.pie(self.valores, labels=self.labels)
        plt.show()
```

```python
grafico_pizza_top_10 = GraficoPizza()
rotulos = lista_nomes
quantidades = lista_bilheteria
grafico_pizza_top_10.criar_grafico(rotulos, quantidades)
grafico_pizza_top_10.exibir_grafico()
```
