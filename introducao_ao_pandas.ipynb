# Pandas

É uma biblioteca (open-source) Python usada para manipular (extrair, analisar, limpar, explorar, sumarizar, ...) conjuntos de dados.

https://pandas.pydata.org/

*Estrutura de dados*: Pandas oferece duas estruturas de dados principais: **Series** e **DataFrame**. Uma Series é um array unidimensional que pode conter qualquer tipo de dados, enquanto um DataFrame é uma estrutura de dados tabular bidimensional que consiste em linhas e colunas.

*Manipulação de dados*: Pandas oferece muitas ferramentas para manipular dados, incluindo filtragem, seleção, ordenação, agregação e transformação. Isso permite que você trabalhe facilmente com dados para realizar tarefas comuns de análise de dados.

*Limpeza de dados*: Pandas oferece uma ampla gama de ferramentas para limpeza de dados, incluindo preenchimento de valores nulos, remoção de duplicatas e correção de erros de dados.

*Visualização de dados*: Pandas integra-se bem com a biblioteca de visualização de dados do Python, Matplotlib, permitindo que você crie visualizações poderosas diretamente de seus dados.

*Importação e exportação de dados*: Pandas pode ler e gravar dados em vários formatos, incluindo CSV, Excel, SQL, JSON e muito mais.

Integração com outras bibliotecas Python: Pandas é uma biblioteca muito popular e é frequentemente usada em conjunto com outras bibliotecas Python, como NumPy, Scikit-Learn e TensorFlow, entre outras.

Definindo uma classe Grafico genérica

```python
import numpy as np
import matplotlib.pyplot as plt

class Grafico:
    def __init__(self):
        self.x = None
        self.y = None

    def set_eixo_x(self,x):
      self.x = x

    def set_eixo_y(self,y):
      self.y =y

    def criar_eixo_x(self, inicio, fim, quantidade):
        self.x = np.linspace(inicio, fim, quantidade)

    def calcular_y(self, funcao):
        self.y = funcao(self.x)

    def configura_limites_dos_eixos(self, limite_inferior_x=None, limite_superior_x=None, limite_inferior_y=None, limite_superior_y=None):
        if ((limite_inferior_x is not None) and (limite_superior_x is not None) and (limite_inferior_y is not None) and (limite_superior_y is not None)):
            plt.xlim([limite_inferior_x, limite_superior_x])
            plt.ylim([limite_inferior_y, limite_superior_y])

    def configura_titulo_e_rotulos(self, titulo=None, rotulo_x=None, rotulo_y=None):
        if titulo is not None:
            plt.title(titulo)
        if rotulo_x is not None:
            plt.xlabel(rotulo_x)
        if rotulo_y is not None:
            plt.ylabel(rotulo_y)

    def exibir_gráfico(self):
        plt.plot(self.x, self.y)
        plt.show()
```

```python
# Exemplo de uso da classe Gráfico
grafico_temperaturas = Grafico()
grafico_temperaturas.set_eixo_x(x=meses)
grafico_temperaturas.set_eixo_y(y=temperaturas)
grafico_temperaturas.configura_titulo_e_rotulos(titulo='Temperaturas em São Paulo - 2022', rotulo_x='Meses', rotulo_y='Temperatura (ºC)')
grafico_temperaturas.exibir_gráfico()
```

### As principais propriedades de um DataFrame no Pandas são:

*index*: os rótulos das linhas do DataFrame. Pode ser um rótulo único para cada linha ou pode ser um MultiIndex que contém vários rótulos para cada linha.

*columns*: os rótulos das colunas do DataFrame. Pode ser um rótulo único para cada coluna ou pode ser um MultiIndex que contém vários rótulos para cada coluna.

*values*: uma matriz que contém os dados do DataFrame. É uma matriz NumPy bidimensional que contém todos os valores do DataFrame.

*shape*: uma tupla que indica o número de linhas e colunas do DataFrame.

*size*: o número total de elementos no DataFrame (número de células).

*dtypes*: uma série que contém os tipos de dados de cada coluna.

*axes*: uma lista de objetos que descrevem os rótulos das linhas e colunas.

*ndim*: o número de dimensões do DataFrame.

*empty*: um valor booleano que indica se o DataFrame está vazio.

```python
print(f'df.index: {df.index}')
print('')
print(f'df.columns: {df.columns}')
print('')
print(f'df.values: {df.values}')
print('')
print(f'df.shape: {df.shape}')
print('')
print(f'df.size: {df.size}')
print('')
print(f'df.dtypes: {df.dtypes}')
print('')
print(f'df.axes: {df.axes}')
print('')
print(f'df.ndim: {df.ndim}')
print('')
print(f'df.empty: {df.empty}')
```

### As principais métodos de um DataFrame no Pandas são:

*head()*: um método que retorna as primeiras n linhas do DataFrame (por padrão, n = 5).

*tail()*: um método que retorna as últimas n linhas do DataFrame (por padrão, n = 5).

*info()*: exibe um resumo conciso do DataFrame, incluindo informações sobre as colunas e seus tipos de dados.

*describe()*: calcula estatísticas descritivas para todas as colunas do DataFrame, incluindo a contagem, média, desvio padrão, mínimo, máximo e quartis.

*shape*: retorna uma tupla com o número de linhas e colunas do DataFrame.

*dropna()*: remove todas as linhas que contêm valores ausentes (NaN).

*fillna()*: preenche os valores ausentes (NaN) com um valor específico.

*groupby()*: agrupa as linhas do DataFrame com base em uma ou mais colunas e aplica uma função de agregação a cada grupo.

*sort_values()*: classifica as linhas do DataFrame com base em uma ou mais colunas.

apply(): aplica uma função a cada elemento de uma coluna ou linha do DataFrame.
