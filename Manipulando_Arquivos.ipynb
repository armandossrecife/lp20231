## Abre um arquivo .txt em modo de leitura

```python
# Operações básicas em arquivos usando Python
# https://docs.python.org/3/tutorial/inputoutput.html#reading-and-writing-files

# abre o arquivo para leitura
my_file = open('teste.txt', mode="r")
```

```python
# Nome do arquivo
# Modo de acesso ao arquivo
# Verifica status do arquivo

nome_arquivo = my_file.name
modo_arquivo = my_file.mode
status_arquivo = my_file.closed

print(f'Nome: {nome_arquivo}')
print(f'Modo: {modo_arquivo}')
print(f'Status: {status_arquivo}')
```

## Ler o conteúdo de um arquivo .txt

```python
conteudo_arquivo = my_file.read()
my_file.close() # libera o arquivo para outras operações

file_size = len(conteudo_arquivo)

if (file_size > 0):
  print(conteudo_arquivo)
else:
  print("Arquivo vazio!")
```

# Exemplo de manipulação de arquivo com PDF

Manipulando PDFs com o fpdf
https://pyfpdf.readthedocs.io/en/latest/

```python
# Exporta um arquivo texto para pdf
!pip install fpdf
```

```python
import datetime

data_hora_atual = datetime.datetime.now()
print(data_hora_atual)
```

```python
# Import the class FPDF from module fpdf
# Add a page
# Set the font
# Insert a cell and provide the text
# Save the pdf with “.pdf” extension

from fpdf import FPDF
 
# save FPDF() class into a
# variable pdf
pdf = FPDF()
# Add a page
pdf.add_page()
# set style and size of font
# that you want in the pdf
pdf.set_font("Arial", size = 15)
# create a cell
conteudo1 = "Laboratório de Programação - " + str(data_hora_atual)
pdf.cell(200, 10, txt = conteudo1, ln = 1, align = 'C')
# add another cell
pdf.cell(200, 10, txt = "Criando um arquivo PDF usando Python e FPDF.", ln = 2, align = 'C')
# add another cell
pdf.cell(200, 10, txt = "Linha 3", ln = 3, align = 'L')
# add another cell
pdf.cell(200, 10, txt = "Linha 4", ln = 4, align = 'L')
pdf.cell(200, 10, txt = "Linha 5", ln = 5, align = 'C')
# add another cell
pdf.cell(200, 10, txt = "Linha 6", ln = 6, align = 'R')
# add another cell
pdf.cell(200, 10, txt = "Linha 7", ln = 7, align = 'R')

# save the pdf with name .pdf
pdf.output("my_file.pdf")  
```

# Exemplo de manipulação de arquivos PPT

https://python-pptx.readthedocs.io/en/latest/index.html

```python
!pip install python-pptx
```

```python
from pptx import Presentation

def cria_slide_principal(apresentacao, titulo="Arquivos de Computador"):
    slide_layout = apresentacao.slide_layouts[0]
    slide = apresentacao.slides.add_slide(slide_layout)
    title = slide.shapes.title
    title.text = titulo

def cria_slide(apresentacao, titulo, conteudo):     
    slide_layout = apresentacao.slide_layouts[1]
    slide = apresentacao.slides.add_slide(slide_layout)
    title = slide.shapes.title
    title.text = titulo
    content = slide.placeholders[1]
    content.text = conteudo

def salva_apresentacao(apresentacao, nome="apresentacao.pptx"):
    try:
        apresentacao.save(nome)
        print("Apresentação criada com sucesso!")
    except Exception as ex:
        print(f"Erro ao gerar o arquivo - {str(ex)}")

def main_ppt():
    presentation = Presentation()
    cria_slide_principal(apresentacao=presentation, titulo="Arquivos de Computador")

    lista_conteudo = []
    lista_conteudo.append(("Introdução", "Os arquivos de computador são a base da organização e armazenamento de informações digitais."))
    lista_conteudo.append(("Tipos de Arquivos", "Existem diferentes tipos de arquivos, como documentos de texto, planilhas, imagens, vídeos e muito mais."))
    lista_conteudo.append(("Sistemas de Arquivos", "Os sistemas de arquivos são responsáveis pela organização e gerenciamento dos arquivos em um dispositivo de armazenamento."))
    
    for i in range(len(lista_conteudo)):
      cria_slide(apresentacao=presentation, titulo=lista_conteudo[i][0], conteudo=lista_conteudo[i][1])
    salva_apresentacao(apresentacao=presentation, nome="apresentacao.pptx")
```

```python
main_ppt()
```
